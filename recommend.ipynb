```python
users_df = pd.read_csv('datasets/users_data.csv')
ans_df = pd.read_csv('datasets/answers_data.csv')
quest_df = pd.read_csv('datasets/questions_data.csv')
# rating_df = pd.read_csv('datasets/rating.csv')
```

```python
merged_dataset = pd.concat([users_df, quest_df], axis=1)
merged_dataset.head()
```

#Colaborative based Filtering Based Recommendation System 

```python
print(merged_dataset['rating '])
```

```python
filter_data = (merged_dataset[['userId', 'city','caseFor','rating ']])
print(filter_data.head())
```

```python
# Create user-item matrix
user_item_matrix = merged_dataset.pivot_table(index='userId', columns='caseFor', values='rating ').fillna(0)
```

```python
print(user_item_matrix)
```

```python
# Calculate similarity
item_similarity = cosine_similarity(user_item_matrix.T)
print(item_similarity)
```

```python
# Create a dictionary to map each caseFor to its average rating
caseFor_ratings = merged_dataset.groupby('caseFor')['rating '].mean().to_dict()
caseFor_ratings
```

```python
# Recommendation function
def recommend_items(caseFor, caseFor_ratings):
    rating = caseFor_ratings.get(caseFor)
    if rating is None:
        return "CaseFor not found"
    recommended_items = [key for key, value in caseFor_ratings.items() if value == rating and key != caseFor]
    return recommended_items
```

```python
# Test the recommendation engine
caseFor_input = 'Bike'  # User input
recommended_items = recommend_items(caseFor_input, caseFor_ratings)
print("Recommended items for", caseFor_input)
print(recommended_items)
```
